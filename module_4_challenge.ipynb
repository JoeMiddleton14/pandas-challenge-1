## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['subtotal'] = df['unit_price'] * df['qty']

# Print the DataFrame to verify the new column
print(df)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
over_50_pounds_rate = 7  
under_50_pounds_rate = 10  

# Create a new column named 'shipping_price' using conditional logic
df['shipping_price'] = df['unit_weight'].apply(lambda x: over_50_pounds_rate * x if x > 50 else under_50_pounds_rate * x)

# Print the DataFrame to verify the new column
print(df)
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

sales_tax_rate = 0.0925

# Create a new column named 'total_price' that calculates the total price
df['total_price'] = df['subtotal'] + df['shipping_price'] * sales_tax_rate

# Print the DataFrame to verify the new column
print(df)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']

# Print the DataFrame to verify the new column
print(df)
```

```python
# Create a column for the profit of
# each line using line cost and line price
df['line_price'] = df['unit_price'] * df['qty']

df['profit'] = df['line_price'] - df['line_cost']

# Print the DataFrame to verify the new column
print(df)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
# Filter the DataFrame to include only orders from the top 5 clients
top_five_client_ids = [33615, 66037, 46820, 38378, 24741]

# Filter the DataFrame to include only orders from the top 5 clients
top_five_clients_orders = df[df['client_id'].isin(top_five_client_ids)]

# Calculate the total spending for each of the top 5 clients
total_spending_by_client = top_five_clients_orders.groupby('client_id')['total_price'].sum()

total_spending_by_client_currency = total_spending_by_client.map('${:,.2f}'.format)

# Print the total spending for each of the top 5 clients in a currency format
print("Total spending for each of the top 5 clients:")
print(total_spending_by_client_currency)
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
top_five_clients_orders = df[df['client_id'].isin(top_five_client_ids)]

# Calculate the total units purchased, total shipping price, total revenue, and total profit for each client
summary_df = top_five_clients_orders.groupby('client_id').agg(
    total_units_purchased=('qty', 'sum'),
    total_shipping_price=('shipping_price', 'sum'),
    total_revenue=('total_price', 'sum'),
    total_profit=('profit', 'sum')
)

# Format the total revenue, shipping price, and profit columns as dollar amounts
summary_df['total_revenue'] = summary_df['total_revenue'].map('${:,.2f}'.format)
summary_df['total_shipping_price'] = summary_df['total_shipping_price'].map('${:,.2f}'.format)
summary_df['total_profit'] = summary_df['total_profit'].map('${:,.2f}'.format)

# Sort the summary DataFrame by total profit
summary_df = summary_df.sort_values(by='total_profit', ascending=False)

# Print the summary DataFrame
print(summary_df)

```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

summary_df['total_revenue'] = summary_df['total_revenue'].str.replace('$', '').str.replace(',', '')
summary_df['total_shipping_price'] = summary_df['total_shipping_price'].str.replace('$', '').str.replace(',', '')
summary_df['total_profit'] = summary_df['total_profit'].str.replace('$', '').str.replace(',', '')

summary_df['total_revenue'] = pd.to_numeric(summary_df[('total_revenue', '')]) / 1000000
summary_df['total_shipping_price'] = pd.to_numeric(summary_df[('total_shipping_price', '')]) / 1000  # Adjusted to divide by 1000 for thousands
summary_df['total_profit'] = pd.to_numeric(summary_df[('total_profit', '')]) / 1000000

# Rename the columns for presentation
summary_df = summary_df.rename(columns={
    'total_units_purchased': 'Total Units Purchased',
    ('total_shipping_price', ''): 'Total Shipping Price (Thousands)',
    ('total_revenue', ''): 'Total Revenue (Millions)',
    ('total_profit', ''): 'Total Profit (Millions)'
})

# Format the total revenue, shipping price, and profit columns as million-dollar amounts
summary_df['Total Revenue (Millions)'] = summary_df['Total Revenue (Millions)'].map('${:,.2f}M'.format)
summary_df['Total Shipping Price (Thousands)'] = summary_df['Total Shipping Price (Thousands)'].map('${:,.2f}K'.format)  # Adjusted to display in thousands (K)
summary_df['Total Profit (Millions)'] = summary_df['Total Profit (Millions)'].map('${:,.2f}M'.format)

# Print the formatted and renamed summary DataFrame
print(summary_df)
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
summary_df['total_profit'] = pd.to_numeric(summary_df['total_profit'].str.replace('$', '').str.replace(',', ''))
summary_df = summary_df.sort_values(by='total_profit', ascending=False)

# Convert total revenue, shipping price, and total profit to millions of dollars
summary_df['total_revenue'] = summary_df['total_revenue'] / 1_000_000
summary_df['total_shipping_price'] = summary_df['total_shipping_price'] / 1_000_000
summary_df['total_profit'] = summary_df['total_profit'] / 1_000_000

# Format the total revenue, shipping price, and profit columns as million-dollar amounts
summary_df['total_revenue'] = summary_df['total_revenue'].map('${:,.2f}M'.format)
summary_df['total_shipping_price'] = summary_df['total_shipping_price'].map('${:,.2f}M'.format)
summary_df['total_profit'] = summary_df['total_profit'].map('${:,.2f}M'.format)

# Print the sorted summary DataFrame
print(summary_df)
```
